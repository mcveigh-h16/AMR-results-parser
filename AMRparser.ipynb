### Read in CuratedAMRFamilies.txt file.

The file can be editted prior to running the script but be sure to save it with the same file name and in the same directory

```python
GeneList = (r'CuratedAMRFamilies.txt') 
genelist_df = pd.read_csv(GeneList, sep='\t', index_col=None, low_memory=False, header=None, names=["genesymbol"])
genelist_df = set(genelist_df['genesymbol'].tolist())
pattern = '|'.join(genelist_df)
```

```python
import re
def pattern_searcher(search_str:str, search_list:str):

    search_obj = re.search(search_list, search_str)
    if search_obj :
        return_str = search_str[search_obj.start(): search_obj.end()]
    else:
        return_str = np.nan
    return return_str
```

```python
AMRresults['allele'] = AMRresults.gene.apply(lambda x: pattern_searcher(search_str=x, search_list=pattern))
#AMRresults
```

```python
AMRresults.loc[AMRresults.gene == AMRresults.allele, "allele"] = np.nan
```

### Final table to save and use with gbench to update your records

```python
AMRresults['gene'],AMRresults['allele'] = np.where(~pd.isnull(AMRresults['allele']), (AMRresults['allele'],AMRresults['gene']),(AMRresults['gene'],AMRresults['allele']))
AMRresults
```

### Check for table rows where the Targe Length = RefSeq length AND %Cov = %Iden
Empty dataframe returns indicates none are found

```python
ExactMatch = AMRresults.loc[(AMRresults['Target length'] == AMRresults['RefSeq length']) & (AMRresults['%Cov'] == AMRresults['%Iden'])]
ExactMatch
```

### Check for partial match scenario where Target Length = RefSeq Length but %Cov not equal to %Iden of vice-versa

Empty dataframe returns indicates none are found

```python
LengthMatch = AMRresults.loc[(AMRresults['Target length'] == AMRresults['RefSeq length']) & (AMRresults['%Cov'] != AMRresults['%Iden'])]
IdentMatch = AMRresults.loc[(AMRresults['Target length'] != AMRresults['RefSeq length']) & (AMRresults['%Cov'] == AMRresults['%Iden'])]
NotMatch = [LengthMatch, IdentMatch]
NotExactMatch = pd.concat(NotMatch)
NotExactMatch
```
